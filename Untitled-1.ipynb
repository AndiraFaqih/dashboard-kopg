```python
import pandas as pd
import os

# --- Definisi Variabel Anda ---
DATA_PATH = os.path.join("data", "KINERJA PERBANKAN.xlsx")
SHEET_NAME = "SUMMARY"  # Nama sheet yang ingin diolah
OUTPUT_FILENAME = "kinerja_perbankan_summary.csv" # Nama file output CSV

# 1. Baca data dari sheet Excel tertentu
try:
    # Membaca data dari file Excel dan sheet yang ditentukan
    df = pd.read_excel(DATA_PATH, sheet_name=SHEET_NAME)
    print(f"✅ Berhasil membaca sheet '{SHEET_NAME}' dari file '{DATA_PATH}'")

    # 2. Simpan DataFrame ke format CSV
    # index=False agar kolom index bawaan pandas tidak ikut tersimpan di CSV
    df.to_csv(OUTPUT_FILENAME, index=False, encoding='utf-8')
    print(f"✅ Berhasil menyimpan data ke file '{OUTPUT_FILENAME}'")

except FileNotFoundError:
    print(f"❌ ERROR: File tidak ditemukan di jalur: {DATA_PATH}")
except ValueError as e:
    if "Worksheet named" in str(e):
        print(f"❌ ERROR: Sheet bernama '{SHEET_NAME}' tidak ditemukan dalam file Excel.")
    else:
        print(f"❌ ERROR: Terjadi kesalahan saat membaca file Excel: {e}")
except Exception as e:
    print(f"❌ ERROR: Terjadi kesalahan yang tidak terduga: {e}")
```
